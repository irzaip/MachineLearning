```python
#conda install -c anaconda graphviz 
# or pip install graphviz

from sklearn import tree
```

```python
#importing from csv
import csv
bod = []
sex = []
with open('bdims.csv', 'rb') as f:
    reader = csv.reader(f)
    for row in reader:
        bod.append(row[0:3])   #full 0:23
        sex.append(row[24])


#check same length
len(bod),len(sex)
print bod[0]
```

```python
#clear first field

bc=bod.pop(0)
sx=sex.pop(0)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(bod,sex)
```

```python
prediction = clf.predict([['42', '33.2', '34']])
print prediction
```
